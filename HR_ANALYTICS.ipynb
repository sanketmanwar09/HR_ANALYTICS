## Average Attrition rate Department wise

```python
avg_attrition_rate = df1.groupby('Department')['Attrition'].apply(lambda x: (x == 'Yes').mean() * 100)
avg_attrition_rate
```

```python
plt.figure(figsize=(10, 6))
avg_attrition_rate.plot(kind='bar', color='blue')
plt.title('Average Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
df = df.rename(columns={'Employee ID' :'Employee Id'})
df.head()
```

```python
df1 = df1.rename(columns={'EmployeeNumber' :'Employee Id'})
df1.head()
```

```python
merged_df = pd.merge(df, df1, on='Employee Id', how='inner')

merged_df.info()
```

## Attrition rate Vs Monthly income stats

```python
Attrition_MI = merged_df.groupby('Attrition')['MonthlyIncome'].mean()
Attrition_MI
```

```python
plt.figure(figsize=(4, 3))
plt.pie(Attrition_MI, labels=Attrition_MI.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'salmon'])
plt.title('Distribution of Monthly Income by Attrition')
plt.axis('equal')  
plt.show()
```

## Average working years for each Department

```python
average_departments_working_years= merged_df.groupby('Department')['TotalWorkingYears'].mean()
average_departments_working_years
```

```python
plt.figure(figsize=(10, 6))
bars = average_departments_working_years.plot(kind='bar', color='green')
plt.title('Average Total Working Years by Department')
plt.xlabel('Department')
plt.ylabel('Average Total Working Years')
plt.xticks(rotation=45, ha='right')

# Add value labels on the bars
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), round(bar.get_height(), 2),
             ha='center', va='bottom')

plt.tight_layout()

# Display the plot
plt.show()
```

## Job Role Vs Work life balance

```python
merged_df['WorkLifeBalance'].head()
```

```python
work_life_balance_by_job_role = merged_df.groupby('JobRole')['WorkLifeBalance'].count()
work_life_balance_by_job_role 
```

```python
plt.figure(figsize=(10, 6))
bars = work_life_balance_by_job_role.plot(kind='bar', color='teal')
plt.title('Work-Life Balance Ratings by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), bar.get_height(),
             ha='center', va='bottom')

plt.tight_layout()

plt.show()
```
